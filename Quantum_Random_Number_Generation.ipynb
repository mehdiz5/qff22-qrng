## 2 Implement the ideas that you came up with in the above paragraph by coding a quantum circuit in Qiskit and execute it on a simulator qasm_simulator

```python
from qiskit import QuantumCircuit
import numpy as np
from qiskit import QuantumRegister
from qiskit import ClassicalRegister
from qiskit.visualization import plot_bloch_vector
import numpy as np
from qiskit.providers.aer import  QasmSimulator
```

```python
from qiskit import Aer, transpile
from qiskit.visualization import plot_bloch_multivector

def execute(circuit, shots):
    backend = Aer.get_backend("qasm_simulator")
    results = backend.run(transpile(circuit, backend), shots = shots,memory=True).result()
    return results.get_memory()
```

```python
# creating the circuit 
qreg = QuantumRegister(1)

qc = QuantumCircuit(qreg)
qc.h(0)
# putting the QC  Qbits in a superposition state

qc.measure_all()

# Generating the random sequence
seq = "".join(execute(qc,5000))
seq
```
